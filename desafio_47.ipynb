```python
import pandas as pd
import matplotlib.pyplot as plt

# Crie um DataFrame com os dados fornecidos
tabela = pd.read_excel("tabela_alunos.xlsx")
display(tabela)

```

```python
df = pd.DataFrame(tabela)

saiu = df[df['Atividade'] == 'Saiu']['Nome'].count()
ficou = df[df['Atividade'] == 'Entrou']['Nome'].count()

labels = ['Saiu', 'Ficou']
sizes = [saiu, ficou]
colors = ['#ff9999', '#66b3ff']
explode = (0.1, 0)  

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')  
plt.title('Porcentagem de Pessoas que Saíram e Ficaram')

plt.show()
```
